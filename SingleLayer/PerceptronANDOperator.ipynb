```python
import numpy as np

inputs = np.array([[0,0], [0,1], [1,0], [1,1]]) ## Creating Matrix
outputs = np.array([0,0,0,1])
weights = np.array([0.0, 0.0])
learning_rate = 0.1

print(inputs.shape) ## Displays Number of Rows and Column
print(outputs.shape)

def step_function(sum):
    if (sum >= 1):
        return 1
    return 0 

## Calculate Output i.e inputs * weights
def calculate_output(instance):
    s = instance.dot(weights)
    return step_function(s)

fun_output = calculate_output(np.array([[1,1]]))
print(f'Function Output before Training is: {fun_output}')


## Train Function (Updates the weight i.e to achieve 100%)
def train():
    total_error = 1 ## Execute untill the error is zero
    while(total_error != 0):
        total_error = 0
        for i in range(len(outputs)):
            prediction = calculate_output(inputs[i])
            error = abs(outputs[i] - prediction)
            total_error += error
            if error > 0:
                for j in range(len(weights)):
                    weights[j] = weights[j] + (learning_rate * inputs[i,j] * error)
                    print('weights Updated: ' + str(weights[j]))
        print('Total Error:  ' + str(total_error))  
         
train()  
after_training_output = calculate_output(np.array([[1,1]]))    
print(f'Function Output After Training is: {after_training_output}')    
```
